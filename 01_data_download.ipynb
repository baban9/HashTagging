```python
from file_downloader import downloader 
import os 
import itertools
from collections import Counter
import pandas as pd 
import time
```

```python
insta_id = os.environ['INSTA_ID']
insta_pass = os.environ['INSTA_PASS']

insta_handler = downloader(insta_id, insta_pass)
```

```python
%cd insta_data_folder/
```

```python
folder_path = "text_files/"
hash_combination = []
file_count = 0
FILTER_THRS = 0.1
hashtag = "fashion"
completed = set()
num_posts = 10
```

```python
try:
    while len(completed) < 20:
        insta_handler.get_files(hashtag, n=num_posts)
        for files in os.listdir(folder_path):
            loc = folder_path + files 
            hash_list = []
            with open(loc, "r", encoding="utf8") as handler:
                data = handler.read() 
            handler.close() 
            for token in data.split():
                if token.startswith("#"):
                    if token.count("#") > 1:
                        token_hashtags = " #".join(token.split("#")).split()
                        for tok_ in token_hashtags:
                            if len(tok_.strip()[1:]) > 1:
                                hash_list.append(tok_.strip())
                    else:
                        if len(token[1:]) > 1:
                            hash_list.append(token)
            hash_combination.extend(list(itertools.product(hash_list,hash_list)))
            file_count += 1

        pair_counter = Counter([x for x in hash_combination if x[0] != x[1]]) 
        results = sorted([[v, k[0], k[1]] for k, v in pair_counter.items() if v > file_count*FILTER_THRS]) 
        cache_ = [] 
        clean_results = []
        for value, left_, right_ in results:
            if (left_, right_) not in cache_ or (right_, left_) not in cache_:
                cache_.append(tuple([left_, right_]))
                cache_.append(tuple([right_, left_]))
                clean_results.append([value, left_, right_])  

        df = pd.DataFrame(clean_results, columns=['weight', "left", 'right'])
        df.to_csv(f"hashtag.csv", index=False)
        extracted = set(df['left'].to_list()).union(set(df['right'].to_list()))
        hashtags = extracted-completed
        completed.add(hashtag) 
        hashtag = hashtags.pop()[1:]  
        print(".", end=" ") 
        time.sleep(10)
except:
    pass

with open("completed_hashtags.txt", "a") as f:
    for c in completed:
        f.write("#"+c)
        f.write("\n")
f.close()    
```
